### Memorization and retrieval

```python
from rbias.constants import C0_NS, C1_NS, M1_NS, C0_CS, C1_CS, M1_CS, C0_M0_ACS, C0_M1_ACS, C1_MO_ACS
from rbias.process import compute_pr
from rbias.viz import multi_bar_plots
```

```python
def compute_pr(res_file, cm_map):
    df = pd.read_csv(res_file)
    df["output"] = df["output"].apply(lambda x: list(map(int, x.strip("[]").split(" "))))
    ndf = df.groupby(["task", "trial_no"])["output"].apply(lambda x: np.concatenate(x.to_list()).tolist())
    res = []
    for task_id in range(TASKS):
        out = np.array(ndf.loc[task_id].to_list())
        rel_c, rel_i = out[:, cm_map[task_id][0]],out[:, cm_map[task_id][1]]
        rel = out[:, cm_map[task_id][0] + cm_map[task_id][1]]
        avg = rel.mean() if rel.size > 0 else 0
        c_avg = rel_c.mean() if rel_c.size > 0 else 0
        i_avg = rel_i.mean() if rel_i.size > 0 else 0
        res.append((avg, c_avg, i_avg))
    return np.array(res)
```

#### Compute $P_r$ to check perfomrance of intervention that changes and doesn't change.

1. Intervene on $B$ and check $A \to B$ - Task 0
2. Intervene on $B$ and check $A \to B$. Intervene on $C$ and check $A \to B$ - Task 1

This is in rchar dataset only.

```python
def compute_m_std(out, task_id, cm_map):
    data = [out[:,f] for i, f in enumerate(cm_map[task_id])]
    row = []
    for d in data:
        avg, sterr = d.mean(), d.std()/np.sqrt(15)
        row.append((avg, sterr))
    return row

def compute_pr_for_inter_type(res_file, cm_map): # interventions which change and don't change the graph
    df = pd.read_csv(res_file)
    df["output"] = df["output"].apply(lambda x: list(map(int, x.strip("[]").split(" "))))
    ndf = df.groupby(["task", "trial_no"])["output"].apply(lambda x: np.concatenate(x.to_list()).tolist())
    res = []
    for task_id in range(TASKS):
        out = np.array(ndf.loc[task_id].to_list())
        row = compute_m_std(out, task_id, cm_map)
        res.append(row)
    return res
```

```python
def compute_pr_for_inter_type(res_file, cm_map): # interventions which change and don't change the graph
    df = pd.read_csv(res_file)
    df["output"] = df["output"].apply(lambda x: list(map(int, x.strip("[]").split(" "))))
    ndf = df.groupby(["task", "trial_no"])["output"].apply(lambda x: np.concatenate(x.to_list()).tolist())
    res = []
    for task_id in range(TASKS):
        out = np.array(ndf.loc[task_id].to_list())
        obv, nobv = out[:, cm_map[task_id][0]], out[:, cm_map[task_id][1]]
        avg_obv, std_obv = obv.mean(), obv.std()/np.sqrt(15)
        avg_nobv, std_nobv = nobv.mean(), nobv.std()/np.sqrt(15)
        res.append([(avg_obv, std_obv), (avg_nobv, std_nobv)])
    return res
    
```

```python
res = []
REL_PRES_ABS_MAP_NO_CHANGE = {
    0: ([1, 3], [1], [3]),
    1: ([2], [5], [8]),
    2: ([1], [4], [7])
}

REL_PRES_ABS_MAP_CHANGE = {
    0: ([0, 2], [0], [2]),
    1: ([0, 1], [4, 3], [6, 7]),
    2: ([0, 2], [5, 3], [6, 8])
}

REL_PRES_ABS_MAP = {
    0: ([1, 3], [0, 2]),
    1: ([2, 5, 8], [0, 1, 4, 6, 3, 7]),
    2: ([1, 4, 7], [0, 2, 5, 6, 3, 8])
}
def compute_m_std(out, task_id):
    data = [out[:,f] for i, f in enumerate(REL_PRES_ABS_MAP_NO_CHANGE[task_id])]
    row = []
    for d in data:
        avg, sterr = d.mean(), d.std()/np.sqrt(15)
        row.append((avg, sterr))
    return row

def compute_pr_for_rel_pres_absent(res_file, cm_map): # interventions which change and don't change the graph
    df = pd.read_csv(res_file)
    df["output"] = df["output"].apply(lambda x: list(map(int, x.strip("[]").split(" "))))
    ndf = df.groupby(["task", "trial_no"])["output"].apply(lambda x: np.concatenate(x.to_list()).tolist())
    res = []
    for task_id in range(TASKS):
        out = np.array(ndf.loc[task_id].to_list())
        out = np.array(ndf.loc[task_id].to_list())
        row = compute_m_std(out, task_id)
        res.append(row)
    return res
    
```
